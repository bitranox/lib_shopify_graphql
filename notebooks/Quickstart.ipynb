## 5) Python API: partial updates and the UNSET sentinel

`ProductUpdate` is a partial update. A field you never name stays `UNSET` and is not sent to Shopify at all, which is what keeps an update from clearing fields you did not mention. Setting a field to `None` is the explicit way to clear it.

```python
from lib_shopify_graphql import UNSET, ProductStatus, ProductUpdate

update = ProductUpdate(title="New Product Title", status=ProductStatus.ACTIVE)

print(f"title:            {update.title!r}")
print(f"status:           {update.status}")
print(f"description_html: {'UNSET, so it is not sent' if update.description_html is UNSET else update.description_html}")
print(f"vendor:           {'UNSET, so it is not sent' if update.vendor is UNSET else update.vendor}")
```

## 6) Python API: the error hierarchy

Every failure the library raises descends from `ShopifyError`, so a caller can catch that one type or narrow to a specific case.

```python
from lib_shopify_graphql import (
    AuthenticationError,
    GraphQLError,
    ProductNotFoundError,
    ShopifyError,
)

for err in (AuthenticationError, ProductNotFoundError, GraphQLError):
    print(f"{err.__name__:<22} is a ShopifyError: {issubclass(err, ShopifyError)}")
```
